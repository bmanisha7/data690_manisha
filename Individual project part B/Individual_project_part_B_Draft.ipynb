```python
#Importing libraries

import pandas as pd
import plotly
import plotly.graph_objects as go
import plotly.express as px
```

```python
#Reading data

df = pd.read_csv('https://raw.githubusercontent.com/bmanisha7/data690_manisha/main/Individual%20project%20part%20B/data/wdi_data%20(2).csv')
df.head(10)
```

```python
#Dropping unwanted rows 

df_new = df.drop(columns=['Unnamed: 0'])
df_new.head()
```

```python
#Taking only particular data of 2019 year

df_2019 = df_new.query("Year == 2019")
df_2019.sample(10)
```

```python
#Filtering only 2019 gdp data

df_2019_gdp = df_2019.query("indicator in['NY.GDP.PCAP.PP.CD']")
df_2019_gdp
```

```python
#Plotting bar graph of 2019 gdp of countries

fig = px.bar(
    data_frame=df_2019_gdp,
    title="Countries by 2019 GDP per Capita",
    labels={"value":"GDP per Capita"},
    x="Country Name",
    y="value",
    color="Country Name"
)

fig.update_xaxes(categoryorder="total descending")


fig.show()
```

```python
#Retrieving only the natural resourses rates of 2019 year

df_2019_nr = df_2019.query("indicator in['NY.GDP.TOTL.RT.ZS']")
df_2019_nr
```

```python
#Plotting the bar graph for the year 2019

fig = px.bar(
    data_frame=df_2019_nr,
    title="Countries by 2019 Total natural resources rents(% of GDP)",
    labels={"value":"GDP per Capita", "Year":"YEAR"},
    x="Country Name",
    y="value",
    color="Country Code"
)

fig.update_xaxes(categoryorder="total descending")


fig.show()
```

```python
#Taking only particular data of 2010 year

df_2010 = df_new.query("Year == 2010")
df_2010.sample(10)
```

```python
#Filtering only total natural resources data of year 2010

df_2010_nr = df_2010.query("indicator in['NY.GDP.TOTL.RT.ZS']")
df_2010_nr
```

```python
#Plotting pir graph for the 2010 Total natural resources rents 

fig = px.pie(
    df_2010_nr,
    values='value', 
    names='Country Name',
    title='Countries by 2010 Total natural resources rents (% of GDP)',
    hover_name='Country Code',
)
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.show()
```

```python
#Getting only total natural resources rents data using indicator

df_nr = df_new.query("indicator == 'NY.GDP.TOTL.RT.ZS'")
df_nr.sample(10)
```

```python
#Plotting line graph for all the years

fig = px.line(
    df_nr,
    x="Year",
    y="value",
    title="Countries by Total natural rents (% of GDP)",
    color='Country Name' 
)
fig.add_annotation(x=2019, y=5,text="Australia",showarrow=False)
fig.add_annotation(x=2019, y=2,text="India", showarrow=False,yshift=10)
fig.add_annotation(x=2019, y=1,text="China", showarrow=False,yshift=10)
fig.add_annotation(x=2019, y=0.5,text="United States", showarrow=False,yshift=10)
fig.add_annotation(x=2019, y=0.0,text="Korea Rep", showarrow=False,yshift=10)

fig.update_layout(showlegend=False)

fig.show()
```
